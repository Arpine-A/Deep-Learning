```python
import os
import shutil
from sklearn.model_selection import train_test_split

# 1. Path to your downloaded dataset
dataset_path = path

# 2. Where to save train/test splits
output_dir = "data"
train_dir = os.path.join(output_dir, "train")
test_dir = os.path.join(output_dir, "test")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# 3. Get list of all image files
all_files = []
for root, dirs, files in os.walk(dataset_path):
    for f in files:
        if f.lower().endswith(('.jpg', '.jpeg', '.png')):
            full_path = os.path.join(root, f)
            all_files.append(full_path)

print("Total images found:", len(all_files))

# 4. Train/test split (80/20)
train_files, test_files = train_test_split(all_files, test_size=0.2, random_state=42)

# 5. Copy files into new folders
def copy_files(file_list, target_folder):
    for file in file_list:
        filename = os.path.basename(file)
        shutil.copy(file, os.path.join(target_folder, filename))

copy_files(train_files, train_dir)
copy_files(test_files, test_dir)

print("Done! Train and test folders created.")
```
